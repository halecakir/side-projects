```python
infile = "record_audio_applist_final_addition.txt"
outfile = "record_audio_applist.txt"
```

```python
apps = {}
with open(infile) as target:
    key = None
    for line in target:
        line = line.strip()
        if line != "":
            if line.startswith("%%"):
                key = line.split("%%")[1]
                apps[key] = []
            else:
                apps[key].append(line.split(".apk")[0])
```

```python
final_list = []
```

```python
for key in apps:
    for i in range(10):
        if len(apps[key]) > 0:
            element = apps[key].pop()
            final_list.append(element)
```

```python
with open(outfile, "w") as target:
    for line in final_list:
        target.write(line + "\n")
```

```python
import csv
import requests
def get_descriptions(id_file, out_file):
    #Get application ids beforehand
    app_infos = {}
    with open(id_file, "r") as target:
        for app_id in target:
            app_infos[app_id.rstrip()] = None

    #Get application descriptions
    base_url = "http://localhost:3000/api/apps/"
    lang = "/?lang=en"
    with open(out_file, "w") as out:
        writer = csv.writer(out)
        writer.writerow(["application_id", "description"])
        for app_id in app_infos:
            response = requests.get(base_url+app_id+lang)
            print(base_url+app_id, end=" ")
            if response.status_code == 200:
                print("SUCCESS")
                if "description" in response.json():
                    writer.writerow([app_id, response.json()["description"]])
            else:
                print("FAIL")

```
